```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib as mpl
import os
```

## Import Waveform Dataframe

```python
skip_waveform_import = False
wave_filepath = ""
while True:
    # Specify the file path directly
    filename = input("Please enter the name of your waveform supply .csv file (Enter 'None' to skip waveform import): ")
    if filename.strip().lower() == 'none':
        skip_waveform_import = True
        break
    folder = input("Please enter the full path (not including the filename) to your .csv file: ")

    if folder:
        waveform_filepath = folder + "\\" + filename
    else: 
        waveform_filepath = filename

    print("\nWaveform Data Filepath: ", waveform_filepath)

    if os.path.isfile(waveform_filepath):
        break
    print("\nThe specified file does not exist. Please check the path and try again.")
```

```python
def import_waveform_csv_to_df(filepath):
    raw_df = pd.read_csv(filepath, delimiter=',', header=10, names=['Time', 'Volts'])
    return raw_df


def plot_waveform(df):
    plt.figure(figsize=(12, 4))
    plt.plot(df['Time'], df['Volts'], linewidth=2)
    plt.title('Voltage Waveform')
    plt.xlabel('Time')
    plt.ylabel('Voltage (V)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()


if not skip_waveform_import:
    waveform_df = import_waveform_csv_to_df(waveform_filepath)
    display(waveform_df)
    plot_waveform(waveform_df)
```
